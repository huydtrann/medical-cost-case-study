```python
import seaborn as sns
import pandas as pd

insurance_data = pd.read_csv('insurance.csv')
insurance_data.head()
```

# How is age related to medical costs?

```python
g = sns.lineplot(data=insurance_data, x='age', y='charges')
g.set_title('Relationship between Age and Medical Costs')
g.set_xlabel('Age')
g.set_ylabel('Medical Costs')
```

The lineplot observes the general trend and provides a sense of how medical costs change as age increases.

# How is number of children related to medical costs?

```python
g = sns.lineplot(data=insurance_data, x='children', y='charges')
g.set_title('Relationship between # of Children and Medical Costs')
g.set_xlabel('Count of Children')
g.set_ylabel('Medical Costs')
```

The lineplot can show the relationship of medical costs for different numbers of children.

# How is the number of people distributed by region?

```python
region_counts = insurance_data['region'].value_counts().reset_index()
region_counts.columns = ['region', 'count']

sns.barplot(data=region_counts, x='region', y='count')

g = sns.barplot(data=region_counts, x='region', y='count')
g.set_title('Number of People Distributed by Region')
g.set_xlabel('Region')
g.set_ylabel('Count')
```

The bar plot displays the number of people in each region, 
showing the distribution across different categories.

# How is the number of people distributed by age?

```python
g = sns.histplot(data=insurance_data, x='age', bins=10)
g.set_title('Distribution of People by Age')
g.set_xlabel('Age')
g.set_ylabel('Count')
```

The histogram shows the frequency distribution of ages, 
providing a visual of how people are spread across different age groups.

# How are the charges distributed?

```python
g = sns.histplot(data=insurance_data, x='charges', bins=30, kde=True)
g.set_title('Distribution of Charges')
g.set_xlabel('Charges')
g.set_ylabel('Count')
```

The histogram shows the distribution of medical charges, 
illustrating how these charges vary in the dataset, 
with lower charges illustrating a higher count compared to higher charges.
